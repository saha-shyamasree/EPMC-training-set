```python
def count(data):
    ner_sent_count = 0
    rel_sent_count = 0
    ner_count = 0
    for pmcid in data:
        annotations = data[pmcid]['annotations']
        for anno in annotations:
            if anno['ner']:
                ner_sent_count += 1
                ner_count += len(anno['ner'])
            if anno['rel']:
                rel_sent_count += 1
    print(f"sents with ner tags: {ner_sent_count}, total ner tags: {ner_count}, sent with rel: {rel_sent_count}")

def compare(before, after):
    for pmcid in before:
        before_annotations = before[pmcid]['annotations']
        after_annotations = after[pmcid]['annotations']
        
        for i in range(len(before_annotations)):
            before_anno = before_annotations[i]
            after_anno = after_annotations[i]
            
#             if before_anno['ner']:
#                 for ner in before_anno['ner']:
#                     if "Hepatic encephalopathy" in ner:
#                         print("=====")
#                         print(before_anno)
#             if after_anno['ner']:
#                 for ner in after_anno['ner']:
#                     if "Hepatic encephalopathy" in ner:
#                         print(after_anno)
#                         print("======")

            if not before_anno['ner'] and not after_anno['ner']:
                continue
            elif not before_anno['ner'] and after_anno['ner']:
                print(after_anno['ner'])
            elif before_anno['ner'] and not after_anno['ner']:
                print(before_anno)
                raise ValueError("after has non")
            else:
                before_set = set(tuple(ann) for ann in before_anno['ner'])
                after_set = set(tuple(ann) for ann in after_anno['ner'])
                
                if not before_set.issubset(after_set):
                    print(f"\nbefore is not subset of after: {before_set}\n{after_set}\n{before_anno}\n{after_anno}\n")

#                 if before_set.difference(after_set):
#                     raise ValueError(f"{before_set.difference(after_set)}")
                if after_set.difference(before_set):
                    print(f"{after_set.difference(before_set)}")
        
        
```
